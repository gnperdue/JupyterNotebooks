```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arr_v = np.load('perfmat1460659054.npy')
```

```python
fig = plt.figure(figsize=(16,16))
plt.imshow(arr_v, cmap=plt.get_cmap('jet'), interpolation='nearest')
fig.savefig('zseg_mat_v_1460659054.pdf')
```

```python
pur_v = np.zeros_like(arr_v)
for i in range(np.shape(arr_v)[0]):
    pur_v[i, :] = arr_v[i, :] / arr_v.sum(axis=1)[i]
```

```python
fig = plt.figure(figsize=(16,16))
plt.imshow(pur_v, cmap=plt.get_cmap('jet'), interpolation='nearest', vmin=0, vmax=1)
fig.savefig('zseg_mat_v_1460659054_pur.pdf')
```

```python
eff_v = np.zeros_like(arr_v)
for i in range(np.shape(arr_v)[0]):
    eff_v[:, i] = arr_v[:, i] / arr_v.sum(axis=0)[i]
```

```python
fig = plt.figure(figsize=(16,16))
plt.imshow(eff_v, cmap=plt.get_cmap('jet'), interpolation='nearest', vmin=0, vmax=1)
fig.savefig('zseg_mat_v_1460659054_eff.pdf')
```
