```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with np.load('models/lminervatriamese_betaprime1457987722.npz') as f:
    param_values = [f['arr_%d' % i] for i in range(len(f.files))]
```

```python
len(param_values)
```

```python
layer_names = {0:'conv-x', 1:'bias-x', 2:'conv-x', 3:'bias-x', 4:'conv-x', 5:'bias-x', 6:'dense-x', 7:'bias-x',
               8:'conv-u', 9:'bias-u', 10:'conv-u', 11:'bias-u', 12:'conv-u', 13:'bias-u', 14:'dense-u', 15:'bias-u',
               16:'conv-v', 17:'bias-v', 18:'conv-v', 19:'bias-v', 20:'conv-v', 21:'bias-v', 22:'dense-v', 23:'bias-v',
               24:'across-dense1', 25:'accros-bias1', 26:'across-dense2', 27: 'accros-bais2'
              }
for i, p in enumerate(param_values):
    print "{:02d}, {:15s}, {}".format(i, layer_names[i], np.shape(p))
```

```python
# look at first x convolution weights
fig = plt.figure(figsize=(18, 30))
gs = plt.GridSpec(4, 3)
for i in range(0, 12):
    ax = plt.subplot(gs[i])
    ax.imshow(param_values[0][i][0], interpolation=None)
    
figname = 'first-x-conv-weights.pdf'
plt.savefig(figname)
```

```python
# look at first u convolution weights
fig = plt.figure(figsize=(18, 30))
gs = plt.GridSpec(4, 3)
for i in range(0, 12):
    ax = plt.subplot(gs[i])
    ax.imshow(param_values[8][i][0], interpolation=None)
    
figname = 'first-u-conv-weights.pdf'
plt.savefig(figname)
```
